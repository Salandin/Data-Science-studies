```python
import numpy as np
from scipy import stats
```

```python
jogadores = [40000, 18000, 12000, 250000, 30000, 140000, 300000, 40000, 800000]
np.mean(jogadores)
```

```python
np.median(jogadores)
```

```python
#criação da variável para geração dos quartis (0, 0.25, 0.50, 0.75, 1)
quartis = np.quantile(jogadores, [0, 0.25, 0.50, 0.75, 1])
quartis
```

```python
#visualização do desvio padrão
np.std(jogadores, ddof=1)
```

```python
#visualização de estatística mais detalhados usando scipy
stats.describe(jogadores)
```
