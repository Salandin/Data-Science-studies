Histograma

```python
# importação
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
base = pd.read_csv('trees.csv')
base.shape
```

```python
base.head()
```

```python
#Craiação do histograma, considerando apenas
#o segundo atributo da base e duas divisões

hist = np.histogram(base.iloc[:,1], bins= 6)
hist
```

```python
#visualização do histograma
plt.hist(base.iloc[:,1], bins=6)
plt.title('Árvores')
plt.xlabel('Frequência')
plt.ylabel('Altura')
```
